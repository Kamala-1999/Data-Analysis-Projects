```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
car_price = pd.read_csv("C:/Users/test/Desktop/Datasets/CarPrice_Assignment (1).csv",index_col = 0)
pd.pandas.set_option('display.max_columns',None)
car_price.head()
```

```python
#Last 5 rows of the data
car_price.tail()
```

```python
#shape
car_price.shape
```

```python
#info
car_price.info()
```

```python
#convet symbols to categorical
car_price['symboling'] = car_price['symboling'].astype('category')
```

```python
#split Car Name and Model into two colmns
car_price['Car'] = car_price['CarName'].str.split(' ').str[0]
```

```python
#split Car Name and Model into two colmns
car_price['Model'] = car_price['CarName'].str.split(' ').str[1]
```

```python
car_price['Model_n'] = car_price['CarName'].str.split(' ').str[2]
```

```python
#Head
car_price.head()
```

```python
#Nan values in Model_n
car_price['Model_n'].isnull().sum()
```

```python
#Joint model_n and Model in a single name
car_price['CarName_1'] = car_price['Model'] + car_price['Model_n']
```

```python
#Head
car_price.head()
```

```python
#fill nan in CarName_1 with corresponding values from Model
car_price['CarName_1'].fillna(car_price['Model'],inplace = True)
```

```python
#Drop Model_n and model
car_price.drop(['Model_n','Model'],axis = 1,inplace = True)
```

```python
#New column Model
car_price['Model'] = car_price['CarName_1']
#drop CarName_1
car_price.drop(['CarName_1'],axis = 1,inplace = True)
```

```python
#fill nan with a space
car_price['Model'].fillna(' ',inplace = True)
```

```python
car_price['Car'] = car_price['Car'] + ' ' + car_price['Model']
```

```python
#Drop Model and carName
car_price.drop(['Model','CarName'],axis = 1,inplace = True)
```

```python
#Head
car_price.head()
```

```python
#all null values
car_price.isnull().sum()
```

```python
#skewness value
car_price.skew()
```

```python
#IQR values
Q1 = car_price['price'].quantile(0.25)
Q3 = car_price['price'].quantile(0.75)
IQR = Q3 - Q1
Lower_bracket = Q1 - 1.5*IQR
Upper_bracket = Q3 + 1.5*IQR
```

```python
#Find outliers in price
car_price[car_price['price'] < Lower_bracket]
car_price[car_price['price'] > Upper_bracket]
       
```

```python
#skewness value of price in graph
sns.distplot(car_price['price'])
plt.axvline(car_price['price'].mean(),color = 'r',linestyle = '--',linewidth = 2)
plt.axvline(car_price['price'].median(),color = 'g',linestyle = '--',linewidth = 2)
```

```python
#Boxplot of price
sns.boxplot(car_price['price'])
```

```python
#square root transformation of price
import scipy.stats as stats
car_price['price_sqrt'] = np.sqrt(car_price['price'])
```

```python
sns.distplot(car_price['price_sqrt'])
plt.axvline(car_price['price_sqrt'].mean(),color = 'r',linestyle = '--',linewidth = 2)
plt.axvline(car_price['price_sqrt'].median(),color = 'g',linestyle = '--',linewidth = 2)
```

```python
#log transformation of price
car_price['price_log'] = np.log(car_price['price'])
```

```python
sns.distplot(car_price['price_log'])
plt.axvline(car_price['price_log'].mean(),color = 'r',linestyle = '--',linewidth = 2)
plt.axvline(car_price['price_log'].median(),color = 'g',linestyle = '--',linewidth = 2)
```

```python
#Box cox transformation of price
car_price['price'] = stats.boxcox(car_price['price'])[0]
```

```python
sns.distplot(car_price['price'])
plt.axvline(car_price['price'].mean(),color = 'r',linestyle = '--',linewidth = 2)
plt.axvline(car_price['price'].median(),color = 'g',linestyle = '--',linewidth = 2)
```

```python
#log transformation of horse power
car_price['horsepower'] = np.log(car_price['horsepower'])
```

```python
sns.distplot(car_price['horsepower'])
plt.axvline(car_price['horsepower'].mean(),color = 'r',linestyle = '--',linewidth = 2)
plt.axvline(car_price['horsepower'].median(),color = 'g',linestyle = '--',linewidth = 2)
```
