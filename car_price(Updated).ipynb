```python
#new dataframe with only numeric values
df_numeric = df.select_dtypes(include=['int64','float64'])
```

```python
#distplot of every column in a single plot
for i in df_numeric.columns:
    sns.distplot(df_numeric[i])
    plt.axvline(df_numeric[i].mean(), color='r', linestyle='--')
    plt.axvline(df_numeric[i].median(), color='g', linestyle='--')
    plt.show()
```

```python
numericals = df.select_dtypes(include=['int64','float64']).columns
```

```python
sns.set_palette('mako')
```

```python
#all distribution plots in a single plot
_,ax = plt.subplots(5,3,figsize=(18,22))
for i in range(len(numericals)):
    sns.histplot(data=df[numericals[i]],ax=ax[i//3][i%3],kde=True)
```

```python
#new dataframe with only categorical values
categorical = df.select_dtypes(include=['category','object'])
categorical
```

```python
#box cox transformation of compressionratio
import scipy.stats as stats
df['compressionratio'] = stats.boxcox(df['compressionratio'])[0]
```

```python
sns.distplot(df['compressionratio'])
plt.axvline(df['compressionratio'].mean(), color='r', linestyle='--')
plt.axvline(df['compressionratio'].median(), color='g', linestyle='--')
plt.show()
```
